```python
import matplotlib.pyplot as plt
```

## ラベル

```python
a = [1, 2, 3, 4, 5]
b = [1, 4, 9, 16, 25]
# データはデータフレームでもnumpyでもリストでもOK

plt.plot(a, b)
plt.title("line graph") ##グラフのタイトル
plt.xlabel("AAA") ##x軸のラベル
plt.ylabel("BBB") ##y軸のラベル
plt.xticks([1, 2, 4, 5]) ##x軸の目盛り
plt.yticks([1, 4, 9, 16, 25]) ##y軸の目盛り
plt.xlim(0, 8) ##x軸の範囲
plt.ylim(0, 30) ##y軸の範囲
plt.legend(["God of graph"]) ##凡例
plt.show() ##グラフを表示(省略可)
```

## 色、スタイル

```python
c = [1,2,3]
d = [4,5,6]
e = [3,6,9]

plt.plot(c, d, color = "red", linestyle = "--")
plt.plot(c, e, color = "blue", linestyle = "-.")
## colorで色を指定、linestyleで線のスタイルを指定
```

## サブプロット

```python
plt.figure(figsize = (10, 4)) ##グラフのサイズを指定

## plt.subplot(行数(縦), 列数(横), グラフの番号)
plt.subplot(1,2,1) ##1つ目のグラフ
plt.plot([1,2,3], [4,5,6]) ##変数なしで直接リストを入れて折れ線グラフを描写
plt.title("1st graph")

plt.subplot(1,2,2) ##2つ目のグラフ
plt.bar(["A", "B", "C"], [3, 7, 2]) 
plt.title("2nd graph")

plt.tight_layout() ##グラフ同士が重ならないように調整
```
