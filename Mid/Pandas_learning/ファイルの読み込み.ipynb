### まずデータを確認して、区切り字(sep)、読み飛ばすべきデータ行の行数(skiprows)、列ラベルの有無(header)を確認する

```python
csv_in = "wholesale_customers.csv" # csvのパス
df = pd.read_csv(csv_in, sep=",", skiprows=13, header=0) # csvの読み込み
# pd.read_csv(パス, sep=区切り文字の指定, skiprows=スキップする行数の指定, header=ヘッダー行数)
df
```
