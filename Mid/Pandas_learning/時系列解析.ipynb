```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "data/hakusan_elec.csv"
df = pd.read_csv(data, skiprows=1)
```

```python
print(df.shape)
print(df.info())
display(df.head())
```

```python
df["Date"] = pd.to_datetime(df["Date"]) ## Date列をdatetime型に変換
print(df.info())
```

```python
df = df.set_index("Date")  ## Date列をインデックスに設定
display(df.head())
```

```python
## 日付情報の取得(いずれも先頭3行を表示)
print(df.index[:3])
print(df.index.year[:3])  ## 年
print(df.index.month[:3])  ## 月
print(df.index.day[:3])    ## 日
print(df.index.date[:3])    ## 日付(年、月、日) ndarray形式
print(df.index.dayofyear[:3])  ## 年内の日数(1月1日=1, 12月31日=365)
print(df.index.dayofweek[:3])  ## 曜日(0=月曜日, 6=日曜日)
print(df.index.quarter[:3])  ## 四半期(1=1月～3月, 2=4月～6月, 3=7月～9月, 4=10月～12月)
```

```python
df_rolling7 = df.rolling(7).mean()  ## 7日間の移動平均
display(df_rolling7.head(10))
```

```python
df_resample_w = df.resample("W").mean()  ## 週ごとの平均
## オーバーラップしない一定幅の期間を設定
## D:日 , W:週, M:月, QE:四半期, YE:年 h:時間, min:分, S:秒
display(df_resample_w.head(10))
```

```python
df["day_of_week"] = df.index.dayofweek  ## 曜日を列に追加
df_wday_ave = df.groupby("day_of_week").mean()  ## 曜日ごとに平均を集計
display(df_wday_ave)
```
