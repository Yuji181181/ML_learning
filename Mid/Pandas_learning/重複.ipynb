```python
import pandas as pd
import numpy as np
```

```python
data = "data/wholesale_customers-with-errors.csv"
df = pd.read_csv(data,skiprows=13)
```

```python
print(df.shape)
print(df.info())
display(df.head())
```

```python
display(df[df.duplicated(keep=False)]) ## 重複データの確認
## keep=False にすると重複している行全てがTrueになる
```

```python
df = df.drop_duplicates().reset_index(drop=True)  ## 重複データの削除
print(df.shape)
display(df.head())
```
