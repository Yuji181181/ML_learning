```python
import pandas as pd
import seaborn as sns
```

```python
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [500, 600, 750, 850, 900],
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture'],
    'Ad_Spend': [100, 200, 300, 400, 500],
    'Values': [7, 8, 9, 5, 4],
    'Region': ['North', 'South', 'East', 'West', 'North']
}
df = pd.DataFrame(data)
df
```

# 可視化のテーマ設定

```python
sns.set_theme(style="darkgrid") #テーマの設定
```

```python
sns.lineplot(data=df, x="Year", y="Sales")
```

# 可視化のカラーパレット設定

```python
sns.set_palette("Set2") #パレットの設定
```

```python
sns.lineplot(data=df, x="Year", y="Sales")
```
