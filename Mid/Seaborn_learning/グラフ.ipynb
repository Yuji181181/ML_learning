```python
import pandas as pd
import seaborn as sns
```

```python
data = {
    'Year': [2020, 2021, 2022, 2023, 2024],
    'Sales': [500, 600, 750, 850, 900],
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture'],
    'Ad_Spend': [100, 200, 300, 400, 500],
    'Values': [7, 8, 9, 5, 4],
    'Region': ['North', 'South', 'East', 'West', 'North']
}
df = pd.DataFrame(data)
df
```

# 折れ線グラフ

```python
sns.lineplot(data=df, x='Year', y='Sales')
```

# 棒グラフ

```python
sns.barplot(data=df, x='Year', y='Sales')
```

# ヒストグラム

```python
sns.histplot(df['Sales'],bins=4,kde=True)
## bins: 棒の数]
## kde: 確率密度関数の表示
```

# 散布図

```python
sns.scatterplot(data=df, x='Ad_Spend', y='Sales')
```

# 箱ひげ図

```python
sns.boxplot(data=df, x='Category', y='Values')
```

# ヒートマップ

```python
corr = df[['Sales', 'Ad_Spend',"Values"]].corr() # 相関係数
corr ## 2次元データ
```

```python
sns.heatmap(corr,annot=True,fmt=".2f",cmap="coolwarm") # 相関係数のヒートマップ
# 値は２次元データ
# annot: 数値の表示
# fmt: 数値の表示形式
# cmap: 色の指定
```

# カウントプロット

```python
sns.countplot(data=df, x='Region')
```

# ペアプロット

```python
sns.pairplot(df[['Sales', 'Ad_Spend', 'Values']])
```
