```python
import matplotlib.pyplot as plt
import numpy as np
```

# １番

```python
plt.hist(np.random.binomial(n=100, p=0.8, size=1000))
```

# ２番

```python
plt.hist(np.random.poisson(lam=10, size=1000))
```

# ３番

```python
plt.hist(np.random.randn(1000) )
```

# ４番

```python
plt.hist(np.random.normal(loc=-2, scale=1.4, size=1000) )
```

# ５番

```python
plt.hist(np.random.exponential(scale=1/1.3, size=1000))
```

# ６番

```python
plt.hist(np.random.rand(1000)*5)
```
