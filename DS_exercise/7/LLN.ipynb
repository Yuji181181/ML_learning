### 大数の法則 Low of large numbers

```python
import numpy as np
import scipy.stats as ss
from matplotlib import pyplot as plt
```

サンプルサイズを10から10000まで10刻みで変えた配列を作る  
Create an array that changes the sample size from 10 to 100000 in 10 steps

```python
sample_size = np.arange(10,10010,10)
print(sample_size[:20])
```

標準正規分布の乱数を生成し、平均値を求める  
Generate the R.V. of the standard normal distribution and  calculate the sample mean 

```python
mean_array = [] 
for i in range(0, len(sample_size)):
    # サンプルサイズがsample_size[i]の標準正規分布の乱数を生成する
    # generate the  R.V. of the standard normal distribution with the sample size sample_size[i]  
    sample = ss.norm.rvs(size = sample_size[i])
    # 生成された乱数から平均値を求め，リストmean_arrayに追加
    # calculate the sample mean
    mean_array.append(np.mean(sample))
```

折れ線グラフにより可視化（横軸：サンプルサイズ"sample_size"，縦軸：平均"mean_array"）  
Draw a line gragh, x-axis is sample size and y-axis is sample mean

```python
plt.plot(sample_size, mean_array)
plt.xlabel("sample size")
plt.ylabel("sample mean")
plt.show()
```
