```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
from scipy.stats import norm
```

```python
data = "screw.csv"
df = pd.read_csv(data, header=None)
df.head()
```

```python
screw_data = df[0].values
```

```python
plt.hist(screw_data, bins=10)
```

```python
sorted_data = np.sort(screw_data)
cumulative = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
plt.plot(sorted_data, cumulative, marker='o')
```

```python
plt.boxplot(screw_data)
```

```python
print(np.mean(screw_data))
print(np.median(screw_data))
print(np.var(screw_data, ddof=1))
print(ss.kurtosis(screw_data))
print(ss.skew(screw_data))
```

```python
lower,upper = ss.norm.interval(0.95, loc=np.mean(screw_data), scale=ss.sem(screw_data))
print(lower, upper)
```
