```python
import pandas as pd
```

```python
csv_in = "data/score.csv"
df = pd.read_csv(csv_in)
df
```

```python
df_class1 = df["class"].head(5)
df_class1
```

```python
new_data = {'id': 170, 'class': 3, 'score1': 80, 'score2': 75}
df2 = pd.DataFrame([new_data])

df = pd.concat([df, df2], ignore_index=True)
df
```

```python
df["ave"] = df["score1"]+df["score2"]/2
df.head(5)
```

```python
df_sort = df.sort_values(by='ave', ascending=False)
df_sort.head(5)
```

```python
df_agg = df.groupby('class').mean()

df_agg2 = df_agg.drop(columns=['id'])
df_agg2
```
