### 例1: 離散型確率分布のKL情報量

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

### 離散型確率分布のKL情報量を求める関数

```python
def kl_divergence(x, y):
    return np.sum(np.where(x != 0, x * np.log(x / y), 0))
```

### 確率分布

```python
p1 = np.array([0.028, 0.121, 0.233, 0.267, 0.2,   0.103, 0.037, 0.009, 0.001])
p2 = np.array([0.014, 0.071, 0.168, 0.24,  0.231, 0.158, 0.079, 0.029, 0.008])
p3 = np.array([0.001, 0.01,  0.044, 0.117, 0.205, 0.246, 0.205, 0.117, 0.044])
```

### 確率分布のグラフ

```python
plt.plot(p1,label="p1")
plt.plot(p2,label='p2')
plt.plot(p3,label='p3')
plt.xlabel('x')
plt.ylabel('probability')
plt.legend()
plt.show()
```

### KL情報量

```python
kld_11 = kl_divergence(p1, p1) # p1とp1間のKL情報量
kld_12 = kl_divergence(p1, p2) # p1とp2間のKL情報量
kld_13 = kl_divergence(p1, p3) # p1とp3間のKL情報量
print("p1-p1: {:.2f}".format(kld_11))
print("p1-p2: {:.2f}".format(kld_12))
print("p1-p3: {:.2f}".format(kld_13))
```
