### 最尤推定(例１)
Maximum Likelihood Estimate (Ex.1)

### 確率分布のパラメータを最尤推定を用いて求める
Find the parameters of the probability distribution using maximum likelihood estimation

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

### 推定対象とするデータを作成 Create data to be estimated
#### パラメータ$\lambda$が2のポアソン分布の乱数を1000個生成
Generate 1000 random numbers with Poisson distribution with parameter $\lambda$

```python
data = ss.poisson.rvs(mu=2, size=1000)
```

### 対数尤度を求める
Find the log-likelihood

```python
para = np.arange(0.1, 5.0, 0.1) # パラメータの推定範囲を0.1から5未満まで0.1刻みとする． The estimated range of parameters is from 0.1 to less than 5 in 0.1 increments.
print(para)
logL = np.array([sum(ss.poisson.logpmf(data, mu=p)) for p in para]) # 各パラメータ値における対数尤度を求める Find the log-likelihood of each parameter value
print(logL)
#print(np.argmax(logL))
mle = para[np.argmax(logL)]# 対数尤度が最大となるパラメータを求める Find the parameter that maximizes the log-likelihood
print(np.argmax(logL))
print("最尤推定値(Maximum likelihood estimated value): {}".format(mle))

plt.plot(para, logL) # 横軸をパラメータ，縦軸を対数尤度とした折れ線グラフをプロット Plot a line graph with parameters on the horizontal axis and log-likelihood on the vertical axis
plt.axvline(x=para[np.argmax(logL)], color='red', linestyle='dotted') # 対数尤度が最大となる点を赤の点線でプロット Plot the point with the highest log-likelihood with a red dotted line
plt.xlabel('parameter')
plt.ylabel('logL')
plt.show()
```
