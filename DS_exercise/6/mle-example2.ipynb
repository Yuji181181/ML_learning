### 最尤推定(例2)
Maximum Likelihood Estimate (Ex.2)

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
data = [0.63, 0.85, 0.61, 1.82, 0.62, 0.24, 1.72, 0.04] # 観測データ　Observation data
para = np.arange(0.1, 5.0, 0.1)
logL = np.array([sum(ss.expon.logpdf(data, scale=1/p)) for p in para]) # 各パラメータ値における対数尤度を求める　Find the log-likelihood of each parameter value

mle = para[np.argmax(logL)]# 対数尤度を最大とするパラメータを求める　Find the parameter that maximizes the log-likelihood
print("最尤推定値(Maximum likelihood estimated value): {:.1f}".format(mle))

plt.plot(para, logL) # 横軸をパラメータ，縦軸を対数尤度とした折れ線グラフをプロット　Plot a line graph with parameters on the horizontal axis and log-likelihood on the vertical axis
plt.axvline(x=para[np.argmax(logL)], color='red', linestyle='dotted') # 対数尤度が最大となる点を赤の点線でプロット　Plot the point with the highest log-likelihood with a red dotted line
plt.xlabel('parameter')
plt.ylabel('logL')
plt.show()
```
