### 例2: 連続型確率分布のKL情報量

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

### 連続型確率分布のKL情報量を求める関数

```python
def kl_divergence_conti(x, y, dt):
    return np.sum(np.where(x != 0, x * np.log(x / y), 0)) * dt
```

### 真の確率分布　(平均1標準偏差1の正規分布)

```python
dt = 0.001 #微小区間幅
x = np.arange(-5, 5, dt) # 確率変数
p_A = ss.norm.pdf(x, loc = 1, scale = 1) # 平均1標準偏差1の正規分布
```

### 真の確率分布のグラフ

```python
plt.plot(x,p_A)
plt.xlabel('x')
plt.ylabel('probability')
plt.title('actual probability distribution')
plt.show()
```

### 各平均の正規分布と真の確率分布間のKL情報量を求める

```python
mean = np.arange(0, 2.1, 0.1)#0から2まで0.1刻みの配列を用意

kl = []
for m in mean:
    p_B = ss.norm.pdf(x, loc = m, scale = 1)
    kld = kl_divergence_conti(p_A, p_B, dt)
    kl.append(kld)
```

### 各平均の正規分布と真の確率分布間のKL情報量のグラフ

```python
plt.plot(mean,kl)
plt.xlabel('m')
plt.ylabel('KL divergence')
plt.show()
```
