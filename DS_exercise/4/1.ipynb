```python
import math
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 80 
K = 10
n = 5
k = 2

comb_daikichi = math.comb(K, k)
comb_other = math.comb(N-K, n-k)
comb_total = math.comb(N, n)

prob = (comb_daikichi * comb_other) / comb_total

print(prob)
```

```python
n = 150  
p = 0.03 

p2 = 1 - ss.binom.cdf(k=k, n=n, p=p)

print(p2)
```

```python
rand = np.random.poisson(lam=5,size=2000)

plt.hist(rand, bins=14)
```
