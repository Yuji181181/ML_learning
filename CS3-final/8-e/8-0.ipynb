```python
import pandas as pd
```

```python
df_name = pd.read_csv('name.csv')
df_mail = pd.read_csv('mail.csv')
```

```python
print(df_name.shape)
print(df_mail.shape)
display(df_name.head())
display(df_mail.head())
```

```python
df_outer = pd.merge(df_name, df_mail, left_on='studentID', right_on='ID', how='outer')
display(df_outer)
```

```python
df_inner = pd.merge(df_name, df_mail, left_on='studentID', right_on='ID', how='inner')
display(df_inner)
```
