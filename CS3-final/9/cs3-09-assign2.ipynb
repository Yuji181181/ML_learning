```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageFilter
```

```python
img_in = 'cat_sample.jpg' 
img = Image.open(img_in)
```

```python
print(img.format, img.size, img.width, img.height, img.mode)
```

```python
plt.imshow(img)
```

```python
data = np.array(img)
print(data.shape)
```

```python
print("R値:", data[600, 500, 0])
print("G値:", data[600, 500, 1])
print("B値:", data[600, 500, 2])
```

```python
img_resized = img.resize((30, 50))
plt.imshow(img_resized)
```

```python
img_rotated = img.rotate(60)
plt.imshow(img_rotated)
```

```python
img_blurred = img.filter(ImageFilter.GaussianBlur(radius=2))
plt.imshow(img_blurred)
```

```python
img_cropped = img.crop((100, 500, 500, 900))
plt.imshow(img_cropped)
```

```python
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_gray.save('cat_gray.jpg', quality=95)
```
