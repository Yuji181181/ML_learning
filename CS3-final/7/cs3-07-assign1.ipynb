```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = "boston_small.csv"
df = pd.read_csv(data,skiprows=8)
```

```python
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_corr = df.corr()
display(df_corr)
```

```python
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
```

```python
print(df.corrwith(df["MEDV"]).idxmin())
```

```python
plt.scatter(df[df.corrwith(df["MEDV"]).idxmin()],df["MEDV"])
```
