```python
import os

# このスクリプトがあるディレクトリ（13-e）を取得
base_dir = os.path.dirname(os.path.abspath(os.getcwd()))

# 13-e/mul フォルダのパス
mul_dir = os.path.join(base_dir, "mul")

# mulフォルダを作成（既に存在してもOK）
os.makedirs(mul_dir, exist_ok=True)

# 九九のファイルを作成
for i in range(1, 10):
    for j in range(1, 10):
        filename = f"{i}x{j}.txt"
        filepath = os.path.join(mul_dir, filename)
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(f"{i} * {j} = {i * j}")

# mulフォルダ内の全txtファイルのパスと中身を表示
for fname in sorted(os.listdir(mul_dir)):
    if fname.endswith(".txt"):
        fpath = os.path.join(mul_dir, fname)
        with open(fpath, "r", encoding="utf-8") as f:
            content = f.read()
        print(f"{fpath}: {content}")
```
