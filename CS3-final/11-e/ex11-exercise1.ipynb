```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
nd_rgb = np.zeros((13, 13, 3), dtype=np.uint8)
nd_rgb[1:4, 1:4, 2] = 255
nd_rgb[5:8, 5:8, 0] = 255
nd_rgb[9:12, 9:12, 1] = 255

image = Image.fromarray(nd_rgb, "RGB")
plt.imshow(image)
```

```python
nd_gray = np.zeros((12, 12), dtype=np.uint8)

nd_gray[2:10, 2:10] = 255
image_gray = Image.fromarray(nd_gray, "L")

plt.imshow(image_gray, cmap='gray')
```
